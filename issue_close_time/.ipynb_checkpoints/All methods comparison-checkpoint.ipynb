```python
from raise_utils.data import DataLoader
from raise_utils.interpret import DODGEInterpreter, ResultsInterpreter
```

```python
datasets = ['cloudstack', 'cocoon', 'deeplearning', 'hadoop', 'hive', 'node', 'ofbiz', 'qpid']
times = ['1 day', '7 days', '14 days', '30 days', '90 days', '180 days', '365 days']
```

```python
for data in datasets:
    for time in times:
        look_for = f"{data}-{time}"
        
        interpreter = ResultsInterpreter([
            f'./nondl/nondl-log/{look_for}-none.txt',
            f'./nondl/nondl-log/{look_for}-weighted.txt',
            f'./nondl/nondl-log/{look_for}-all.txt'
        ]).with_dodge(
            DODGEInterpreter(files=[f'./ghost/log/{look_for}.txt'], 
                             exclude_cols=[-1], 
                             metrics=['d2h', 'pd', 'pf', 'prec']),
            merge_method=lambda r, d: "-".join(r.split("/")[-1].split("-")[:-1]) == d.split("/")[-1].split(".")[0]
        )
        interpreter.interpret()
```
