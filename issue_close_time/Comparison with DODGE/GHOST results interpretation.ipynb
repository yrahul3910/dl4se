```python
from raise_utils.interpret import DODGEInterpreter
import numpy as np


directories = ["1 day", "7 days", "14 days",
               "30 days", "90 days"]
datasets = ["cloudstack", "hadoop", "cocoon",
            "deeplearning", "hive", "node", "ofbiz", "qpid"]
path = "./orig-ghost-log/"

for data in datasets:
    for time in directories:
        try:
            ri = DODGEInterpreter(
                [f"{path}{data}-{time}.txt"], metrics=["d2h", "pd", "pf", "prec", "auc", "f1"])
            results = ri.interpret()[f"{data}-{time}.txt"]
            
            print(f"{data}-{time}")
            print("=" * len(f"{data}-{time}"))
            
            for key in ["pd", "pf", "d2h", "prec"]:
                if key != "d2h":
                    print(key, "-", round(np.median(results[key]), 2), "-", round(0.3 * np.std(results[key]), 2))
                else:
                    print(key, "-", 
                          round(
                              np.sqrt(
                                  (np.median(results["pf"]) ** 2 + 
                                   (1. - np.median(results["pd"])) ** 2) / 2), 2))
            print()
        except: pass
```
